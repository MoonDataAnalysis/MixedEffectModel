## 1.2 随机区组设计（Randomized Block Design）
- 既有固定效应，又有随机效应的简单案例

#### 案例：
- 实验人员记录了九名不同受试者从四种不同类型的凳子上起身所需的力气。

- type 是 实验因子（experimental factor）
- subject 是 区组因子（blocking factor）


```r
library(nlme)
```

- 可视化：每个水平的响应平均值、整体平均值。

```r
plot.design( ergoStool )
```

#### 模型结构


-    也可以写成矩阵形式：


#### contrast 问题
- R 默认使用 treatment contrast，第一个水平被固定作为基准
- 截距项是Type1

```r
contrasts( ergoStool$Type )
```

```r
fm1Stool <- lme(effort ~ Type, data = ergoStool, random = ~ 1 | Subject)
summary( fm1Stool )
```

```r
anova( fm2Stool )
```

- 使用 helmert contrast
- 截距项是4种type的总平均值

```r
options( contrasts = c( factor = "contr.helmert",  ordered = "contr.poly" ) )
contrasts( ergoStool$Type )
```

```r
fm2Stool <- lme(effort ~ Type, data = ergoStool, random = ~ 1 | Subject)
summary( fm2Stool )
```

#### 残差可视化

```r
plot(fm1Stool)
```

```r
plot( fm1Stool, form = resid(., type = "p") ~ fitted(.) | Subject, abline = 0 )
```
