### 线性混合效应（LME）模型 VS 非线性混合效应（NLME）模型

#### 为什么？
- 可解释性（interpretability）：NLME 模型的参数通常具有实际意义，能与研究的实际过程或现象相联系，帮助理解变量之间的内在关系。

- 简约型（parsimony）：NLME 模型在描述数据时，往往使用更少的参数就能达到较好的拟合效果。

- 观测数据范围之外的有效性（validity beyond the observed range of the data）：NLME 模型在对超出观测数据范围的情况进行预测时，表现更为可靠。


- 在选择回归模型来描述响应变量如何随协变量变化时，人们总是可以选择使用多项式模型，这类模型在参数上是线性的。通过增加多项式的次数，可以更紧密地拟合数据。

- 但只是对观测数据内真实函数的近似，且缺乏对潜在机制的理论思考 。

#### 案例：吲哚美辛药代谢动力学
- 六名人类志愿者接受了相同剂量的吲哚美辛静脉推注，并在注射后 15 分钟至 8 小时内测量了 11 次血浆中药物浓度（单位：微克 / 毫升）。


```r
library(nlme)
plot( Indometh ) 
```

- 房室模型（compartment model）

吲哚美辛浓度的机制模型来自一个双房室模型。它把时间 t 时的预期浓度表示为两个指数项的线性组合



如果不考虑组效应





```r
fm1Indom.nls <- nls( conc ~ SSbiexp(time, A1, lrc1, A2, lrc2), data = Indometh )
summary(fm1Indom.nls)
```

```r
plot( fm1Indom.nls, Subject ~ resid(.), abline = 0 )
```

- 为了给每个受试者分别拟合一个双指数模型，从而使个体效应能够纳入参数估计中，我们将模型表示为：



```r
fm1Indom.lis <- nlsList(conc ~ SSbiexp(time, A1, lrc1, A2, lrc2), data = Indometh )
fm1Indom.lis
```

```r
plot( intervals(fm1Indom.lis) )
```

- 为了引入非线性混合效应模型中固定效应和随机效应的概念


6.5 的非线性混合效应版本为



```r
fm1Indom.nlme <- nlme( fm1Indom.lis, random = pdDiag(A1 + lrc1 + A2 + lrc2 ~ 1) )
fm1Indom.nlme
```

```r
fm2Indom.nlme <- update( fm1Indom.nlme, random = pdDiag(A1 + lrc1 + A2 ~ 1) )
```

```r
anova( fm1Indom.nlme, fm2Indom.nlme )
```

- 为了进一步探究保留在 fm2Indom.nlme 模型中的随机效应的方差 - 协方差结构，我们使用一个一般正定的 Ψ 矩阵来更新拟合



```r
fm3Indom.nlme <- update( fm2Indom.nlme, random = A1+lrc1+A2 ~ 1 )
```

```r
fm3Indom.nlme 
```

```r
fm4Indom.nlme <- update( fm3Indom.nlme, random = pdBlocked(list(A1 + lrc1 ~ 1, A2 ~ 1)) )
anova( fm3Indom.nlme, fm4Indom.nlme )
```

```r
plot( fm4Indom.nlme, id = 0.05, adj = -1 )
```

```r
qqnorm( fm4Indom.nlme )
```

```r
plot( augPred(fm4Indom.nlme, level = 0:1) )
```

```r
summary( fm4Indom.nlme )
```

#### 案例： 大豆植株的生长

nlme 在 growth curve data 中的应用


```r
Soybean[1:3, ]
```

```r
plot( Soybean, outer = ~ Year * Variety )
```

```r
fm1Soy.lis <- nlsList( weight ~ SSlogis(Time, Asym, xmid, scal), data = Soybean )
```

```r
fm1Soy.lis
```

```r
fm1Soy.nlme <- nlme( fm1Soy.lis )
fm1Soy.nlme
```

```r
fm2Soy.nlme <- update( fm1Soy.nlme, weights = varPower() )
```

```r
anova( fm1Soy.nlme, fm2Soy.nlme )
```

```r
plot(ranef(fm2Soy.nlme, augFrame = T), form = ~ Year * Variety, layout = c(3,1))
```

```r
soyFix <- fixef( fm2Soy.nlme )
```

```r
options( contrasts = c("contr.treatment", "contr.poly") )
fm3Soy.nlme <- update( fm2Soy.nlme,
                      fixed = Asym + xmid + scal ~ Year,
                      start = c(soyFix[1], 0, 0, soyFix[2], 0, 0, soyFix[3], 0, 0) )
fm3Soy.nlme
```

```r
anova( fm3Soy.nlme )
```

```r
summary( fm3Soy.nlme )
```
